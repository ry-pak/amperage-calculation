```python
'''
расчёт силы тока в цепи
идём построчно, каждую строку превращаем в список, из которого достаём индекс
необходимой величины (ЭДС, R1, R2, R4, p, l, s) для расчёта формулы; затем по индексу достаём значение
этой величины исходя из паттерна написания текста (берём следующий индекс в случае с ЭДС или идём на два индекса вперёд в остальных случаях)
таким образом, код работает для любого текста при условии, что величина ЭДС пишется как 'ЭДС:',
а остальные величины имеют два знака (например '=' и ' ') перед своим значением в тексте (и учитывается только заглавные/ не заглавные буквы соответственно)
'''
i = e = r1 = r2 = r3 = r4 = p = s = l = None

#функция для получения из текста значений R, l, s, p
def find_resistance(x, Rpls_str):
        rpls_position = file_data.splitlines()[x].split(' ').index(Rpls_str)
        rpls = (file_data.splitlines()[x].split(' '))[rpls_position+2]
        return float(rpls)

#функция для расчёта по формуле
def find_amperage(e, r1, r2, r4, p, s, l):
    r3 = (p*l)/s                           #0.00011464968152866241
    i = e/ (((r1 + r2) * r3)/ (r1 + r2 + r3) + r4)
    return i

with open('D:/Pycharm by URFU/home_task1.txt', 'r', encoding = 'utf-8') as file:
    
    file_data = file.read()
    #считаем количесвто строк
    lines = file_data.count('\n')
    
    #идём по каждой строке и ищем в ней необходимые величины
    for x in range(lines+1):
        print('line:', x)
        file_data.splitlines()[x].split(' ')
        print(file_data.splitlines()[x].split(' '))

        if 'ЭДС' in file_data.splitlines()[x]:
            e_position = file_data.splitlines()[x].split(' ').index('ЭДС:')
            print(e_position)
            e = int((file_data.splitlines()[x].split(' '))[e_position+1])
            print("e = ",e)

        if 'R1' in file_data.splitlines()[x]:
            r1 =  find_resistance(x, 'R1')
            print("r1 = ", r1)

        if 'R2' in file_data.splitlines()[x]:
            r2 = find_resistance(x, 'R2')
            print("r2 = ", r2)
        
        if 'R4' in file_data.splitlines()[x]:
            r4 = find_resistance(x, 'R4')
            print("r4 = ", r4)

        if 'p' in file_data.splitlines()[x]:
            p = find_resistance(x, 'p')
            print("p = ", p)

        if 'l' in file_data.splitlines()[x]:
            l = find_resistance(x, 'l')
            print("l = ", l)

        if 's' in file_data.splitlines()[x]:
            s = find_resistance(x, 's')
            print("s = ", s)

        #проверка полученных значений
        print(f"e = {e}, r1 = {r1}, r2 = {r2}, r4 = {r4}, p = {p}, l = {l}, s = {s}")
    
    print(find_amperage(e, r1, r2, r4, p, s, l))             #2.499856699763714


```
